versa, based on user input.

```python
t=int(input("Enter the temperature"))
unit = int(input ("what is the unit of the above temperature ?Enter 1 for fahrenheit and 2 for celsius"))
if unit == 1:
  c = (t-32)*5/9
  print(c,"C")
else:
  f = t*9/5 + 32
  print(f,"F")
```

program to determine if a year is a leap year or not

```python
x = int(input("Enter the year "))
if x%4 == 0:
  if x%100 == 0:
    print("The year is not a leap year")
  else:
    print("The year is a leap year")
elif x%400 == 0:
  print("The year is a leap year")
else:
  print("The year is not a leap year")
```

program that categorizes a student's grade

```python
x = int(input("Enter your grades "))
if x>89 and x<100:
  print("You have an A")
elif x>79 and x<89:
  print("You have a B")
elif x>69 and x<79:
  print("You have a C")
elif x>59 and x<69:
  print("You have a D")
elif x>0 and x<59:
  print("You have an F")
else:
  print("Invalid data")
```

```python
y = int(input("Enter a number "))
for number in range(0,11):
  a = y*number
  print(y,"x",number,"=",a)
```
